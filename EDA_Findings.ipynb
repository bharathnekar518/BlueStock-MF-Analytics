# 📈 BlueStock Mutual Fund Analytics: 10 Key EDA Findings

This document summarizes the core performance, risk, and demographic insights discovered during the Exploratory Data Analysis phase.

---

### 1. Cyclical Bottoming & Structural Recovery
Historical time-series tracking reveals a synchronized macroeconomic bottoming out across all forty mutual fund schemes during early 2022, marking a resilient post-pandemic liquid asset stabilization phase.
* **Supporting Asset Reference:** `outputs/nav_trends.png` 

### 2. High-Growth Breakout Trajectories
The persistent bull market rally of 2023 acted as a major expansion catalyst, driving sharp, upward compounding momentum specifically within Equity-heavy fund classes.
* **Supporting Asset Reference:** `outputs/nav_trends.png` 

### 3. Price Consolidation & Downside Adjustments
Throughout 2024, daily price movements shifted into a classic horizontal consolidation phase characterized by healthy downside corrections before continuing their long-term growth into 2025–2026.
* **Supporting Asset Reference:** `outputs/nav_trends.png` 

### 4. Absolute Asset Management Dominance
Asset Under Management (AUM) allocation logs expose massive institutional concentration, with SBI AMC cementing absolute market dominance by scaling to an unprecedented milestone of Rs. 12.5 Lakh Crores.
* **Supporting Asset Reference:** `outputs/aum_growth.png` 

### 5. Historical Retail Investment Milestones
Monthly Systematic Investment Plan (SIP) configurations show massive linear acceleration over the three-year horizon, culminating in an all-time peak ecosystem inflow of Rs. 31,002 Crores in December 2025.
* **Supporting Asset Reference:** `outputs/sip_inflow_trend.png` 

### 6. Category Inflow Density Concentration
The investment intensity grid map highlights that capital inflows are disproportionately clustered within Large-cap and Flexi-cap equity domains, leaving niche thematic fund classes heavily underserved.
* **Supporting Asset Reference:** `outputs/category_heatmap.png` 

### 7. Primary Core Working Demographics
Age distribution metrics establish that working-class retail participants aged 25–45 control the largest collective block of account volume, forming the true operational engine of the digital platform
* **Supporting Asset Reference:** `outputs/investor_demographics.png` 

### 8. Premium Surreal Capital Surpluses
While younger cohorts maintain a high absolute count of active accounts, box plot dispersion confirms that the highest individual median SIP ticket sizes originate exclusively from the mature 46–60 age tier.
* **Supporting Asset Reference:** `outputs/investor_demographics.png` 

### 9. Urban vs. Rural Geographic Asymmetry
Geographic distribution analytics confirm that Top 30 (T30) urban capital states like Maharashtra and Karnataka capture the overwhelming majority of investment volumes, though Beyond 30 (B30) tier-2 cities represent a steadily growing retail share.
* **Supporting Asset Reference:** `outputs/geographic_distribution.png` 

### 10. Systemic Return Coupling & High Diversification Risk
The statistical return matrix reveals an incredibly high daily performance correlation (coefficients exceeding 0.85) among the top ten equity funds, warning investors of low practical asset diversification across competing fund managers.
* **Supporting Asset Reference:** `outputs/nav_correlation_matrix.png`
