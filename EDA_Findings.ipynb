# 📈 BlueStock Mutual Fund Analytics: 10 Key EDA Findings

This document summarizes the core analytical insights discovered during the Exploratory Data Analysis phase, capturing asset trends, investor demographics, and systemic market behavior from 2022 to 2026.

### 1. Daily NAV Cyclical Recovery (2022 Milestone)
The historical timeline shows a definitive structural bottoming out across all 40 fund schemes early in 2022, marking a highly resilient post-COVID macroeconomic recovery phase as systemic liquid assets stabilization took place.

### 2. High-Growth Momentum During the 2023 Rally
The 2023 market rally acted as a major catalyst for asset value growth, with compounding bullish trends causing significant sharp upward breakout trajectories across Equity-heavy fund schemes.

### 3. Price Stabilization and 2024 Market Corrections
During 2024, the daily NAV trend lines captured a classic consolidation phase. Healthy market downside corrections and pullbacks occurred, which stabilized values before transitioning into the late-stage extended timeline growth of 2025–2026.

### 4. Absolute Market Share Dominance by SBI AMC
Asset Under Management (AUM) analysis reveals massive structural concentration, with SBI AMC establishing absolute market dominance by confidently hitting the benchmark milestone of **Rs. 12.5L Cr.**, vastly outpacing competing fund houses.

### 5. Exponential Growth in Monthly SIP Inflows
Systematic Investment Plan (SIP) metrics show powerful linear momentum from 2022 through 2025, culminating in an all-time peak milestone inflow of **Rs. 31,002 Cr. in December 2025**, proving a massive rise in retail investor trust.

### 6. Asset Class Heatmap Inflow Density Concentration
The category-wise investment intensity grid map highlights that net monthly inflows are heavily concentrated in specific sector buckets—predominantly Large-cap and Flexi-cap Equity domains—leaving niche categories comparatively underserved.

### 7. Core Demographic Driving Investment Capital
Investor age group distribution pie charts show that the largest chunk of capital allocation is heavily driven by working professionals aged **25–45**, highlighting that tech-savvy, mid-career segments form the core of the digital mutual fund ecosystem.

### 8. Premium SIP Ticket Sizes Within Mature Age Bands
While younger cohorts (18–25) account for a high volume of active transaction counts, the box plot distribution proves that the largest median individual SIP ticket sizes (INR amounts) originate from the **46–60** age bracket, indicating higher disposable surpluses.

### 9. Geographic Volume Asymmetry (T30 vs. B30 Split)
Geographic distribution logs reveal that the Top 30 (T30) urban state hubs (such as Maharashtra and Karnataka) continue to control the majority of total investment volume, though Beyond 30 (B30) tier-2 regions are steadily climbing in micro-retail participation.

### 10. High Returns Correlation Across Core Equity Portfolios
The pairwise daily return correlation matrix indicates highly coupled behavior (coefficients > 0.85) among the top 10 selected equity funds. This proves low statistical diversification between competing asset managers, as they tend to share highly similar underlying baseline holdings.
